```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('./Podatki/pn2019.csv', delimiter=';', encoding='mbcs')
data
```

```python
upravne_enote=data.UpravnaEnotaStoritve.unique()
upravne_enote = upravne_enote[:-1]
```

```python
data.columns
```

```python
month=data.DatumPN.str.split(".").str[1].astype('int')
data['month']=month
```

```python
#data_by_month usebuje podatke razdeljene po mesecih od 1 - 12 
data_by_month=[]
for i in range(12):
    data_by_month.append(data[data.month==i+1])
    print(data[data.month==i+1])
```

```python
st_nesrec_v_obcinah_na_mesec = dict()
for obcina in upravne_enote:
    st_nesrec_v_obcinah_na_mesec[obcina]=[]
for obcina in upravne_enote:
    print(obcina+":")
    for month in data_by_month:
        st_nesrec_v_obcinah_na_mesec[obcina].append(month[month.UpravnaEnotaStoritve==obcina].shape[0])
        print(month[month.UpravnaEnotaStoritve==obcina].shape[0])
    print()
```

graf števila nesreč glede na mesec v ljubljani

```python
months=["jan","feb","mar","apr","maj","jun","jul","avg","sep","oct","nov","dec"]
plt.plot(months,st_nesrec_v_obcinah_na_mesec["LJUBLJANA"])
plt.xlabel("Mesec")
plt.ylabel("št. nesreč");
```

v vsakem mesecu posebaj je bilo največ nesreč v ljubljani kar je razvidno iz spodnjega izpisa

```python
for i,month in enumerate(data_by_month):
    print(months[i]+":")
    maximum=0
    obc=""
    for obcina in upravne_enote:
        tmp=month[month.UpravnaEnotaStoritve==obcina].shape[0]
        if(tmp>maximum):
            maximum=tmp
            obc=obcina
    print(obc)
    print()
```
